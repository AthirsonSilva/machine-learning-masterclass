```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tips.csv')
df
```

```python
df['tip'].apply(lambda n: n * 2)
```

```python
df['total_bill'].apply(lambda bill: bill * 2)
```

```python
df.drop('doubled_tip', axis=1)
```

```python
def quality(total_bill: float, tip: float):
  if (tip / total_bill) > 0.15:
    return 'Generous'

  return 'Other'
```

```python
quality(16.99, 1.01)
```

```python
df['quality'] = df[['total_bill', 'tip']].apply(lambda df: 
                                                quality(df['total_bill'], 
                                                        df['tip']), 
                                                        axis=1)

df['quality']
```

```python
df['Quality'] = np.vectorize(quality)(df['total_bill'], df['tip'])
df
```
