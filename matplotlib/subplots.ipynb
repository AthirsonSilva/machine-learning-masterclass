```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = x ** 4
```

```python
a = np.arange(0, 10)
b = a * 3
```

```python
from numpy.lib.npyio import save
fig, axes = plt.subplots(nrows=2, ncols=2)
fig.tight_layout()

axes[0][0].plot(x, y)
axes[0][1].plot(b, a)
axes[1][0].plot(x, b)
axes[1][1].plot(a, y)

fig.subplots_adjust(wspace=.5, hspace=.5)

fig.savefig('four_plots.png')
```

```python
fig1, axes1 = plt.subplots(nrows=3, ncols=1)

axes1[0].plot(x, a)
axes1[1].plot(y, b)
axes1[2].plot(x, y)

fig1.suptitle('Figure level')
fig1.subplots_adjust(wspace=.5, hspace=.75)
```

```python
fig1.savefig('new_subplots.png')
```
