```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./country_table.csv')
df
```

```python
df = df.set_index('Countries')
```

```python
plt.figure(dpi=200)
sns.heatmap(df.drop('Life expectancy', axis=1), 
            linewidths=0.33, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy', axis=1), 
            linewidths=0.33, annot=True, cmap='viridis')
```

```python
plt.figure(dpi=200)
sns.clustermap(df.drop('Life expectancy', axis=1), 
            linewidths=0.33, annot=True, cmap='viridis', col_cluster=False)
```
