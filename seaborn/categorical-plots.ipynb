```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./dm_office_sales.csv')

df.head()
```

```python
plt.figure(figsize=(15, 5),dpi=200)
sns.countplot(data=df, x='level of education', hue='division')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.barplot(data=df, x='level of education', y='salary', 
            estimator=np.mean, ci='sd', hue='division')

plt.legend(bbox_to_anchor=(1.05, 1))
```

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.boxplot(data=df, x='level of education', y='salary')
```

```python
plt.figure(figsize=(12, 4), dpi=100)
sns.boxplot(data=df, x='salary', y='level of education')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.violinplot(data=df, x='salary')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.violinplot(data=df, y='salary', x='division')
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.swarmplot(data=df, y='salary', x='division', size=3)
```

```python
plt.figure(figsize=(10, 4), dpi=200)
sns.boxenplot(data=df, x='salary', y='division')
```
