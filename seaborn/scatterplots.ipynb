```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./dm_office_sales.csv')
df
```

```python
plt.figure(figsize=(12, 4), dpi=200)

sns.scatterplot(x='salary', y='sales', data=df, hue='work experience', 
                size='work experience', alpha=0.5)

plt.savefig('salary_sales_work_experience.png')
```

```python
plt.figure(figsize=(12, 4), dpi=200)

sns.scatterplot(x='salary', y='sales', data=df, hue='training level')

plt.savefig('salary_sales_training_level.png')
```
