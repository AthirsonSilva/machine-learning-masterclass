```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 10)
y = x * x
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, label='X vs Y')
ax.plot(x * 2, y * 2, label='X*2 vs Y*2')

ax.legend(loc=(1, 0.5))
```
