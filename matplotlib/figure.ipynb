```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 20)
y = x ** 4
```

```python
x
```

```python
y
```

```python
plt.plot(x, y)
```

```python
fig = plt.figure()

# large axes
axes1 = fig.add_axes([0, 0, 1, 1])

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('X')
axes1.set_ylabel('y')
axes1.set_title('Bigger picture')

# small axes
axes2 = fig.add_axes([1, 1, 0.5, 0.5])

axes1.plot(x, y)
axes2.plot(x / 2, y / 2)

axes2.set_xlim(0,4)
axes2.set_ylim(0,4000)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed in')

plt.show()
```

```python
fig = plt.figure(dpi=100, figsize=(10, 5))

axe1 = fig.add_axes([0, 0, 1, 1])

axe1.plot(x, y)

fig.savefig('new_figure.png')
```
