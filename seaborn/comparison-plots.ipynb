```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('./StudentsPerformance.csv')
df.tail()
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', shade=True)
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='scatter', alpha=0.3, hue='gender')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='hex')
```

```python
sns.jointplot(data=df, x='math score', y='reading score', kind='kde', hue='gender')
```
